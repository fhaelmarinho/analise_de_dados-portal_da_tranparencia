## Converter as colunas em str para substituir "," por "."

```python
df_viagens['Valor diárias'] = df_viagens['Valor diárias'].astype(str).str.replace(",", ".")
df_viagens['Valor passagens'] = df_viagens['Valor passagens'].astype(str).str.replace(",", ".")
df_viagens['Valor devolução'] = df_viagens['Valor devolução'].astype(str).str.replace(",", ".")
df_viagens['Valor outros gastos'] = df_viagens['Valor outros gastos'].astype(str).str.replace(",", ".")
```

```python
df_viagens.info()
```

## Converter str to float

```python
df_viagens['Valor diárias'] = df_viagens['Valor diárias'].astype(float)
df_viagens['Valor passagens'] = df_viagens['Valor passagens'].astype(float)
df_viagens['Valor devolução'] = df_viagens['Valor devolução'].astype(float)
df_viagens['Valor outros gastos'] = df_viagens['Valor outros gastos'].astype(float)
```

## Despesas totais

```python
df_viagens["Despesas totais"] = df_viagens["Valor diárias"] + df_viagens["Valor passagens"] + df_viagens["Valor devolução"] + df_viagens["Valor outros gastos"]
```

## Cargos que mais gastaram com viagens (Acima de R$ 10 milhões)

```python
gastos_totais_por_cargo = df_viagens.groupby('Cargo')['Despesas totais'].sum().reset_index()
```

```python
gastos_totais_por_cargo[gastos_totais_por_cargo['Despesas totais']> 10_000_000]
```

## Listagem das despesas médias e totais agrupadas por Cargo e destinos frequentes

```python
df_final=(
    df_viagens
    .groupby("Cargo")
    .agg(
        despesa_media=("Despesas totais", "mean"),
        despesas_totais=("Despesas totais", "sum"),
        destino_mais_frequente=("Destinos", pd.Series.mode),
        n_viagens=("Nome", "count")
        )
    .head(5)
    .sort_values(by='n_viagens', ascending=True)
    .reset_index()
    )
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(16,6))
ax.barh(df_final["Cargo"], df_final["n_viagens"], color="#002e50")
fig.suptitle("Viagens por Cargo 2024")
plt.grid(color="gray", linestyle="--", linewidth=0.5)
plt.xlabel("Número de viagens")

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(16,6))
ax.barh(df_final["Cargo"], df_final["despesa_media"], color="#002e50")
fig.suptitle("Despesas médias por viagem")
plt.grid(color="gray", linestyle="--", linewidth=0.5)
plt.xlabel("Despesas médias")

plt.show()
```
